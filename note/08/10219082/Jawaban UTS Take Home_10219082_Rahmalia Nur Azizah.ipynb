# Simple Stats

Data:
$$
X = (4, 7, 5, 1, 3, 6, 4, 8, 5, 7, 2, 5, 6, 3, 9, 6, 8, 5, 7, 4)
$$

1. Write codes to obtain $x$ from $X$ and number $N$ of $x$.
2. Write codes to obtain mean $x_{mean}$, median $x_{med}$, and mode $x_{mod}$ of $x$.
3. Write codes to obtain maximum $x_{max}$, minimum $x_{min}$, and range $x_{range}$ of $x$.
4. Write codes to obtain variance $x_{var}$ and standard deviation $x_{sdev}$ of $x$.

```python
import numpy as np
import math

# Data X
X = [4, 7, 5, 1, 3, 6, 4, 8, 5, 7, 2, 5, 6, 3, 9, 6, 8, 5, 7, 4]

# Mendapatkan list nilai x dari X
for i in range (len(X)):
    x = X[i]
    print ("x"+str(i+1)+":", x)
print( )

# Menghitung banyaknya x pada array X (N)
N = len(X)
print("N: ", N)
```

```python
# Mendapatkan nilai x_mean
sum = 0
for i in range (N):
    x = X[i]
    sum = sum + x

x_mean = sum/N

# Mendapatkan nilai x_med
X_urut = sorted(X)   # mengurtkan nilai x dari terkecil ke terbesar
print("X_urut:", X_urut)
print( )

X_urut = np.array(X_urut)
if N%2 == 0:
    x_med = 1/2*(X_urut[int(N/2)] + X_urut[int(N/2+1)])
else:
    x_med = (X_urut[int((N+1)/2)])

# Mendapatkan nilai x_mod
count = {}
for i in X_urut:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1
        
for x, f in count.items():
    if f == max(count.values()):
        x_mod = x
        
print ("x_mean:", x_mean)
print("x_med:", x_med)
print("x_mod:", x)
```

```python
# Mendapatkan nilai x_max, x_min, x_range
x_max = np.max(X)
x_min = np.min(X)
x_range = x_max - x_min

print("x_max:", x_max)
print("x_min:", x_min)
print("x_range:", x_range)
```

```python
# Mendapatkan x_var dan x_sdev

sum_var = 0
for i in range (N):
    s = (X[i] - x_mean)**2
    sum_var = sum_var + s

x_var = sum_var/(N-1)
x_sdev = math.sqrt(x_var)

print ("x_var:", x_var)
print ("x_sdev:", x_sdev)
```

```python
import math

lower_limit = 0
upper_limit = math.pi
partisi = 10

def f(x):
    return math.sin(x)

# Rectangle left point
def rectl(x0,xn,n):
    h = (xn - x0) / n
    
    integral = f(x0) + f(xn)
    
    for i in range(1,n):
        k = x0 + (i-1)*h
        integral = integral + 2 * f(k)
    
    integral = integral * h/2
    
    return integral 

Luas_rectl = rectl(lower_limit, upper_limit, partisi)
print("Luas area di bawah kurva metode rectangle left: ", Luas_rectl)

# Rectangle mid point
def rectm(x0,xn,n):
    h = (xn - x0) / n
    
    integral = f(x0) + f(xn)
    
    for i in range(1,n):
        k = x0 + (i-0.5)*h
        integral = integral + 2*f(k)
    
    integral = integral * h/2
    
    return integral 

Luas_rectm = rectm(lower_limit, upper_limit, partisi)
print("Luas area di bawah kurva metode rectangle mid: ", Luas_rectm)

# Rectangle right point
def rectr(x0,xn,n):
    h = (xn - x0) / n
    
    integral = f(x0) + f(xn)
    
    for i in range(1,n):
        k = x0 + i*h
        integral = integral + 2 * f(k)
    
    integral = integral * h/2
    
    return integral 

Luas_rectr = rectr(lower_limit, upper_limit, partisi)
print("Luas area di bawah kurva metode rectangle left: ", Luas_rectr)

# Trapezium Method
def trapez(x0,xn,n):
    h = (xn - x0) / n
    
    integral = f(x0) + f(xn)
    
    for i in range(1,n):
        k = x0 + i*h
        integral = integral + 2 * f(k)
    
    integral = integral * h/2
    
    return integral 

Luas_trapez = trapez(lower_limit, upper_limit, partisi)
print("Luas area di bawah kurva metode trapezium: ", Luas_trapez)


# Simpson Method
def simpson(x0,xn,n):
    h = (xn - x0) / n
    
    integral = f(x0) + f(xn)
    
    for i in range(1,n):
        k = x0 + i*h
        
        if i%2 == 0:
            integral = integral + 2 * f(k)
        else:
            integral = integral + 4 * f(k)
    
    integral = integral * h/3
    
    return integral

Luas_simpson = simpson(lower_limit, upper_limit, partisi)
print("Luas area di bawah kurva metode simpson: ", Luas_simpson)
```

```python
def get_a_b(n):
    h = (np.pi/2-0) / n
    x = np.linspace(0, np.pi/2, n+1)
    # Get A
    A = np.zeros((n+1, n+1))
    A[0, 0] = 1
    A[n, n] = -2+4*h**2
    A[n, n-1] = 2
    for i in range(1, n):
        A[i, i-1] = 1
        A[i, i] = -2+4*h**2
        A[i, i+1] = 1

    # Get b
    b = np.zeros(n+1)
    for i in range(1, n+1):
        b[i] = 4*h**2*x[i]
    
    return x, A, b

x = np.pi/2
v = x - np.sin(2*x)

n_s = []
errors = []

for n in range(3, 100, 5):
    x, A, b = get_a_b(n)
    y = np.linalg.solve(A, b)
    n_s.append(n)
    e = v - y[-1]
    errors.append(e)
    
plt.figure(figsize = (10,8))
plt.plot(n_s, errors)
plt.yscale('log')
plt.xlabel('n gird points')
plt.ylabel('errors at x = $\pi/2$')
plt.show()
```
