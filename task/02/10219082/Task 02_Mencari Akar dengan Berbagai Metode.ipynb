# Root Existence Interval
Check existence of root in an interval

```python
# Menentukan Keberadaan Akar (Root Existence)
import math
import matplotlib.pyplot as plt

def f(x):
    y = math.sin(3*x - 0.3*math.pi) + 0.1
    return y

x = [i*0.0  for i in range(0, 100 + 1)]
y = [f(i) for i in x]

xroot = [0, 1]
yroot = [f(i) for i in xroot]

# Mendefinisikan root existence
# Root exist jika yroot[0]*yroot[1]<0
Root_existence = True

if yroot[0]*yroot[1]<0:
    Root_existence = True
else:
    Root_existence = False

plt.grid()
plt.plot(x, y, 'b-', xroot, yroot, 'ro', markerfacecolor='none', ms=10, lw=2)
plt.show()

print("xroot =", xroot)
print("yroot =", yroot)
print("f(x1) f(x2) =", yroot[0] * yroot[1], "< 0")
print("Root existence:", Root_existence)
```


# Scanning Method
Find a root in an interval using scanning method

\begin{equation}
\label{eqn:eq1}
\tag{1}
f(x)= \sin(3x - 0.3\pi) + 0.1
\end{equation}

```python
import math

def f(x):
    y = math.sin(3*x - 0.3*math.pi) + 0.1
    return y

x = [i * 0.01  for i in range(0, 100 + 1)]
y = [f(i) for i in x]

xroot = [0, 1]
yroot = [f(i) for i in xroot]

print("Root existence between", xroot[0], "and", xroot[1], ":", end=' ')
if yroot[0] * yroot[1] < 0:
    print(True)
else:
    print(False)
```

```python
N = 1
xbeg = xroot[0]
xend = xroot[1]
dx = (xend - xbeg) / N


sign = 0
for i in range(N + 1):
    xi = xbeg + i * dx
    xi = round(xi * N) / N
    
    yi = f(xi)
    if yi > 0:
        cur_sign = 1
    else:
        cur_sign = -1
    
    if i == 0:
        sign = cur_sign
    
    if cur_sign * sign < 0:
        xleft = xi - dx
        xright = xi
        break

print("Root exists between", xleft, "and", xright)
```


# Bisection Method
Find a root in an interval using bisection method

\begin{equation}
\label{eqn:eq1}
\tag{1}
f(x)= \sin(3x - 0.3\pi) + 0.1
\end{equation}

```python
import math

def f(x):
    y = math.sin(3*x - 0.3*math.pi) + 0.1
    return y

x = [i * 0.01  for i in range(0, 100 + 1)]
y = [f(i) for i in x]

xroot = [0, 1]
yroot = [f(i) for i in xroot]

print("Root existence between", xroot[0], "and", xroot[1], ":", end=' ')
if yroot[0] * yroot[1] < 0:
    print(True)
else:
    print(False)
```

```python
x1 = xroot[0]
x2 = xroot[1]
x3 = 0.5 * (x1 + x2)
xroot.append(x3)         # menambahkan x3 ke dalam list xroot

eps = 1E-6
dx = 10
i = 2
print("iter root         f(root)")
while dx > eps and i < 100:
    x1 = xroot[i-2]
    x2 = xroot[i-1]
    x3 = xroot[i]
    
    if i == 2:
        print(f"{i-2:04d}", f"{x1:.6E}", f"{f(x1):+.6E}")
        print(f"{i-1:04d}", f"{x2:.6E}", f"{f(x2):+.6E}")
        print(f"{i:04d}", f"{x3:.6E}", f"{f(x3):+.6E}")
   
    f13 = f(x1) * f(x3)
    f23 = f(x2) * f(x3)
     
    if f13 < 0:
        x4 = 0.5 * (x1 + x3)
        xroot[i-2] = x2
        xroot[i-1] = x1
    else:
        x4 = 0.5 * (x2 + x3)
    
    xroot.append(x4)
    dx = abs(f(x4))
    
    print(f"{i:04d}", f"{x4:.6E}", f"{f(x4):+.6E}")
    i += 1
```

```python
print("root =", xroot[-1])
print("f(root) =", f(xroot[-1]))
```

```python
yroot = [f(i) for i in xroot]
```

```python
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, '-', xroot, yroot, 'r.', lw=2)
plt.show()
```

```python
plt.grid()
plt.xlabel("iter")
plt.ylabel("f(x)")
plt.plot(yroot, 'r.-')
plt.show()
```

```python
plt.grid()
plt.xlabel("iter")
plt.ylabel("x")
plt.plot(xroot, 'r.-')
plt.show()
```


# Newton Raphson Method
Find a root in an interval using newton-raphson method

\begin{equation}
\label{eqn:eq1}
\tag{1}
f(x)= \sin(3x - 0.3\pi) + 0.1
\end{equation}

```python
import math

def f(x):
    y = math.sin(3*x - 0.3*math.pi) + 0.1
    return y

x = [i * 0.01  for i in range(0, 100 + 1)]
y = [f(i) for i in x]

def df(x):
    dy = 3*math.cos(3*x - 0.3*math.pi)
    return dy
```

\begin{equation}
\label{eqn:eq2}
\tag{2}
x_{n+1} = x_n - {f(x_n) \over f'(x_n)}
\end{equation}

```python
xroot = []
yroot = []

x0 = 0
y0 = f(x0)

xroot.append(x0)
yroot.append(y0)

i = 1
eps = 1E-6
dx = 10
print("iter root         f(root)")
while dx > eps and i < 20:
    x0 = xroot[i-1]
    if i == 1:
        print(f"{i-1:04d}", f"{x0:.6E}", f"{f(x0):+.6E}")
    
    x1 = x0 - f(x0) / df(x0)
    
    xroot.append(x1)
    dx = abs(f(x1))
    
    print(f"{i:04d}", f"{x1:.6E}", f"{f(x1):+.6E}")
    i += 1
```

```python
print("root =", xroot[-1])
print("f(root) =", f(xroot[-1]))
```

```python
yroot = [f(i) for i in xroot]
```

```python
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, '-', xroot, yroot, 'r.', lw=2)
plt.show()
```

```python
plt.grid()
plt.xlabel("iter")
plt.ylabel("f(x)")
plt.plot(yroot, 'r.-')
plt.show()
```

```python
plt.grid()
plt.xlabel("iter")
plt.ylabel("x")
plt.plot(xroot, 'r.-')
plt.show()
```


# Secant Method
Find a root in an interval using secant method

\begin{equation}
\label{eqn:eq1}
\tag{1}
f(x)= \sin(3x - 0.3\pi) + 0.1
\end{equation}

```python
import math

def f(x):
    y = math.sin(3*x - 0.3*math.pi) + 0.1
    return y

x = [i * 0.01  for i in range(0, 100 + 1)]
y = [f(i) for i in x]
```

\begin{equation}
\label{eqn:eq3}
\tag{3}
x_{n+2} = x_{n+1} - {(x_{n+1}-x_n) f(x_{n+1}) \over f(x_{n+1})-f(x_n))}
\end{equation}

```python
xroot = []
yroot = []

x0 = 0
y0 = f(x0)
xroot.append(x0)
yroot.append(y0)

x1 = 1
y1 = f(x1)
xroot.append(x1)
yroot.append(y1)

i = 2
eps = 1E-6
dx = 10
print("iter root         f(root)")
while dx > eps and i < 20:
    x0 = xroot[i-2]
    x1 = xroot[i-1]
    if i == 2:
        print(f"{i-2:04d}", f"{x0:.6E}", f"{f(x0):+.6E}")
        print(f"{i-1:04d}", f"{x1:.6E}", f"{f(x1):+.6E}")
    
    x2 = x1 - (x1 - x0) * f(x1) / ( f(x1) - f(x0) )
    
    xroot.append(x2)
    dx = abs(f(x2))
    
    print(f"{i:04d}", f"{x2:.6E}", f"{f(x2):+.6E}")
    i += 1
```

```python
print("root =", xroot[-1])
print("f(root) =", f(xroot[-1]))
```

```python
yroot = [f(i) for i in xroot]
```

```python
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, '-', xroot, yroot, 'r.', lw=2)
plt.show()
```

```python
plt.grid()
plt.xlabel("iter")
plt.ylabel("f(x)")
plt.plot(yroot, 'r.-')
plt.show()
```


# Fixed Point Iteration Method
Find a root in an interval using fixed point iteration method

\begin{equation}
\label{eqn:eq1}
\tag{1}
f(x)= \sin(3x - 0.3\pi) + 0.1
\end{equation}

\begin{align}
\label{eqn:eq4}
\tag{4}
0 &= \sin(3x - 0.3\pi) + 0.1 \\
 &= 2\sin\bigg({3 \over 2}x - 0.15\pi \bigg) \cos\bigg({3 \over 2}x - 0.15\pi\bigg) + 0.1 \\
-0.1 &= 2\sin\bigg({3 \over 2}x - 0.15\pi \bigg) \cos\bigg({3 \over 2}x - 0.15\pi\bigg) \\
-{0.05 \over \cos({3 \over 2}x - 0.15\pi)} &= \sin\bigg({3 \over 2}x - 0.15\pi \bigg) \\
2\sin\bigg({3 \over 2}x - 0.15\pi \bigg) &= -{0.05 \over \cos({3 \over 2}x - 0.15\pi)} \\
{3 \over 2}x - 0.15\pi &= \sin^{-1} \bigg [-{0.05 \over \cos({3 \over 2}x - 0.15\pi)} \bigg] \\
{3 \over 2}x &= \sin^{-1} \bigg [-{0.05 \over \cos({3 \over 2}x - 0.15\pi)} \bigg] + 0.15\pi \\
x &= {2 \over 3} \sin^{-1}\bigg [-{0.05 \over \cos({3 \over 2}x - 0.15\pi)} \bigg] + 0.1\pi \\
x &= - {2 \over 3} \sin^{-1}\bigg [{0.05 \over \cos({3 \over 2}x - 0.15\pi)} \bigg] + 0.1\pi
\end{align}

\begin{equation}
\label{eqn:eq5}
\tag{5}
x_{n+1} = - {2 \over 3} \sin^{-1} \bigg[{0.05 \over \cos({3 \over 2}x_{n+1} - 0.15\pi)} \bigg] + 0.1\pi
\end{equation}

```python
import math

def f(x):
    y = math.sin(3*x - 0.3*math.pi) + 0.1
    return y

x = [i * 0.01  for i in range(0, 100 + 1)]
y = [f(i) for i in x]

def g(x):
    a = 0.15 * math.pi
    b = math.cos(1.5 * x - a)
    c = 0.05
    d = -2/3 * math.asin(c / b) + 2 * c * math.pi
    return d
```

```python
xroot = []
yroot = []

x0 = 0
y0 = f(x0)
xroot.append(x0)
yroot.append(y0)

i = 1
eps = 1E-6
dx = 10
print("iter root         f(root)")
while dx > eps and i < 20:
    x0 = xroot[i-2]
    x1 = xroot[i-1]
    if i == 1:
        print(f"{i-1:04d}", f"{x1:.6E}", f"{f(x1):+.6E}")
    
    x1 = g(x0)
    
    xroot.append(x1)
    dx = abs(f(x1))
    
    print(f"{i:04d}", f"{x1:.6E}", f"{f(x1):+.6E}")
    i += 1
```

```python
print("root =", xroot[-1])
print("f(root) =", f(xroot[-1]))
```

```python
yroot = [f(i) for i in xroot]
```

```python
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, '-', xroot, yroot, 'r.', lw=2)
plt.show()
```


# False Position Method
Find a root in an interval using false position method

\begin{equation}
\label{eqn:eq1}
\tag{1}
f(x)= \sin(3x - 0.3\pi) + 0.1
\end{equation}

```python
import math

def f(x):
    y = math.sin(3*x - 0.3*math.pi) + 0.1
    return y

x = [i * 0.01  for i in range(0, 100 + 1)]
y = [f(i) for i in x]
```

\begin{equation}
\label{eqn:eq6}
\tag{6}
x_{n+2} = {x_n f(x_{n+1}) - x_{n+1}f(x_n) \over {f(x_{n+1})-f(x_n)}}
\end{equation}

```python
xroot = []
yroot = []

x0 = 0
y0 = f(x0)
xroot.append(x0)
yroot.append(y0)

x1 = 1
y1 = f(x1)
xroot.append(x1)
yroot.append(y1)

i = 2
eps = 1E-6
dx = 10
print("iter root         f(root)")
while dx > eps and i < 20:
    x0 = xroot[i-2]
    x1 = xroot[i-1]
    if i == 2:
        print(f"{i-2:04d}", f"{x0:.6E}", f"{f(x0):+.6E}")
        print(f"{i-1:04d}", f"{x1:.6E}", f"{f(x1):+.6E}")
    
    x2 =  ( x0 * f(x1) - x1 * f(x0) ) / ( f(x1) - f(x0) )
    
    xroot.append(x2)
    dx = abs(f(x2))
    
    print(f"{i:04d}", f"{x2:.6E}", f"{f(x2):+.6E}")
    i += 1
```


# Steffensen Method
Find a root in an interval using steffensen's method

\begin{equation}
\label{eqn:eq1}
\tag{1}
f(x)= \sin(3x - 0.3\pi) + 0.1
\end{equation}

```python
import math

def f(x):
    y = math.sin(3*x - 0.3*math.pi) + 0.1
    return y

x = [i * 0.01  for i in range(-1, 100 + 1)]
y = [f(i) for i in x]
```

\begin{equation}
\label{eqn:eq7}
\tag{7}
x_{n+1} = x_n - {f^2(x_n) \over {f(x_n + f(x_n)) - f(x_n)}}
\end{equation}

```python
xroot = []
yroot = []

x0 = 0.2
y0 = f(x0)
xroot.append(x0)
yroot.append(y0)

i = 1
eps = 1E-6
dx = 10
print("iter root         f(root)")
while dx > eps and i < 20:
    x0 = xroot[i-1]
    if i == 1:
        print(f"{i-1:04d}", f"{x0:.6E}", f"{f(x0):+.6E}")
    
    x1 = x0 - ( f(x0) * f(x0) ) / ( f(x0 + f(x0)) - f(x0) )
    
    xroot.append(x1)
    dx = abs(f(x1))
    
    print(f"{i:04d}", f"{x1:.6E}", f"{f(x1):+.6E}")
    i += 1
```

```python
print("root =", xroot[-1])
print("f(root) =", f(xroot[-1]))
```

```python
yroot = [f(i) for i in xroot]
```

```python
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, '-', xroot, yroot, 'r.', lw=2)
plt.show()
```
